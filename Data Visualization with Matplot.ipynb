# Data Visualization
### What is Data Visualization?
* In today's business world, with so much information collected through data analysis, we need to find a way to draw a picture of the data so that we can interpret it.
* Data visualization gives us a clear idea of what the information means by providing a visual context through maps or graphs.
* This makes the data more natural for the human mind to understand and therefore makes it easier to identify trends, patterns, and outliers in large data sets.
* In short, data visualization is the representation of data or information in a graph, chart, or other visual format.

### The Importance of Data Visualization
* The main purpose is to visualize data and statistics, to gain insights, and to interpret screens.
* Data visualization is used for data cleaning, exploring data structure, identifying outliers and normal groups, detecting trends and clusters, identifying local patterns, evaluating model outputs, and summarizing results.
* It facilitates analyses such as machine learning and predictive analysis, and these analyses can later serve as useful visualizations for presentation.
* Whether you work in finance, marketing, technology, design, or any other field, you need to visualize data. This fact demonstrates the importance of data visualization.

### Most Commonly Used Python Libraries for Data Visualization
1. Matplotlib: The basic visualization library.
2. Seaborn: A statistical data visualization library for Python.
3. Plotly: An interactive data visualization library for Python.

```python
t = np.arange(0.0, 5.0, 0.1)

plt.plot(t, t, '^:k', t, t**2, 'bo', t, t**3, 'g^')
plt.show()
```

```python
rawdata = {
    'a' : np.arange(50),
    'c' : np.random.randint(0,50,50),
    'd' : np.random.randn(50)
}

rawdata['b'] = rawdata['a'] + (10 * np.random.randn(50))
rawdata['d'] = np.abs(rawdata['d']) * 100

plt.scatter('a', 'b', c = 'c', s = 'd', data = rawdata)
plt.xlabel('data in column a')
plt.ylabel('data in column b')
plt.show()
```

```python
names = ['A','B','C','D']
numbers = [5,25,60,50]

plt.figure(figsize = (9,3))

plt.subplot(131)
plt.bar(names, numbers)

plt.subplot(132)
plt.scatter(names, numbers)

plt.subplot(133)
plt.plot(names, numbers)

plt.suptitle('Different Graphs')
plt.show()
```

```python
nu1 = 100
sigma1 = 15
x = nu1 + (sigma1 * np.random.randn(10000))

nu2 = 110
sigma2 = 10
y = nu2 + (sigma2 * np.random.randn(10000))

n, bins, patches = plt.hist(x, 100, density = 1, facecolor = 'b', alpha = 0.25)
n, bins, patches = plt.hist(y, 100, density = 1, facecolor = 'g', alpha = 0.25)

plt.xlabel('Data')
plt.ylabel('Possibilities')
plt.title(r'Data Histogram: $\mu_1 = 100, \ \mu_2 = 110$')
plt.annotate('Top of the Gauss curve', xy = (110, 0.0425), xytext = (130, 0.045), 
             arrowprops = dict(facecolor = 'black', shrink = 0.1))
plt.text(75, 0.025, r'$\nu_1 = 100, \ \sigma_1 = 15$')
plt.text(120, 0.035, r'$\nu_2 = 110, \ \sigma_2 = 10$')
plt.axis([40, 160, 0, 0.05])
plt.grid(True)
plt.show()
```
