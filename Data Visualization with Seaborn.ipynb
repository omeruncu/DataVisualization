# Seaborn Library
### Seaborn is a statistical data visualization library for Python based on the Matplotlib library.
### Seaborn provides a high-level interface for users to create statistical visualizations.
### The functions of the Seaborn library are as follows:
* It is a data set-based application programming interface provided to examine the relationships between multiple variables in detail.
* It supports observing categorical variables and performing statistical operations on them.
* It supports multiple graphs that allow for creating complex visualizations.
* It offers visualizations for comparing subsets with single-variable and pair-variable visualizations.
* There is a wide range of colors that make visualizations more understandable.
### Differences between the Seaborn and Matplotlib libraries:
* Matplotlib can be customized; however, making the plot more understandable and visually appealing is quite difficult.
* Seaborn offers different solutions to users with its color palette, interfaces, and variety of graphs.
* Seaborn provides ease of use when working with the Pandas library.

# Olympic History Dataset

```python
os.chdir('....')
olympicData = pd.read_csv('olymData.csv')
```

```python
olympicData.rename(columns ={
    'isim' : 'name', 
    'cinsiyet' : 'gender', 
    'yas' : 'age', 
    'boy' : 'height', 
    'kilo' : 'weight', 
    'takim' : 'team', 
    'uok' : 'noc',    
    'yil' : 'year', 
    'sezon' : 'season', 
    'sehir' : 'city', 
    'spor' : 'sport', 
    'etkinlik' : 'event', 
    'madalya' : 'medal'}, inplace = True)
```

```python
olympicData.head()
```

# Scatter plot

* Scatter plot is used to determine the relationship between two different variables.

```python
def scatterGraph():
    sns.scatterplot(x = 'height', y = 'weight', data = olympicData)
    plt.title('Height & Weight Relation - White Theme')
    plt.show()
```

```python
sns.set_style('whitegrid')
scatterGraph()
```

```python
sns.set_style('darkgrid')
scatterGraph()
```

# Scatter plot - Categorical

```python
sns.scatterplot(x = 'height', y = 'weight', hue = 'medal', data = olympicData)
plt.title('Height and weight distribution according to medals')
plt.show()                
```

```python
sns.set_style('whitegrid')

sns.scatterplot(x = 'height', y = 'weight', hue = 'gender', style = 'medal',
                size = 'age', sizes = (1, 100), data = olympicData)

plt.xlabel('Height of athletes')
plt.xlabel('Weight of athletes')
plt.title('Height and weight distribution according to genders')
plt.show()        
```

```python
sns.relplot(x = 'height', y = 'weight', hue = 'gender', style = 'season', col = 'medal',
                size = 'age', sizes = (1, 100), data = olympicData)
```

# Scatter plot - Linear Regression

```python
sns.set_style('white')
sns.regplot(x = 'height', y = 'weight', marker = '+', scatter_kws={'alpha': 0.2}, 
            data = olympicData)

plt.title('Height and weight distribution')
plt.show() 
```

# Color Palette

```python
sns.scatterplot(x = 'height', y = 'weight', hue = 'medal', palette = 'rocket', data = olympicData)
plt.title('Height and weight distribution according to medals')
plt.show()      
```

# Line Graph

* A line graph is a type of chart that shows the information of a series of data points connected by straight line segments.
* The chart is generally used to visually represent the values of data over a specific time interval - time series - and is drawn chronologically.
* It is also used to visualize measurements that are dependent on time.

```python
sns.set_style('white')

sns.lineplot(x = 'height', y = 'weight', data = olympicData)
plt.title('Height and weight distribution')
plt.show()      
```

# Line Graph - Categorical

```python
sns.set_style('white')

sns.lineplot(x = 'height', y = 'weight', hue = 'gender', data = olympicData)
plt.title('Height and weight distribution according to gender')
plt.show() 
```

```python
sns.set_style('white')

sns.lineplot(x = 'height', y = 'weight', hue = 'season', data = olympicData)
plt.title('Height and weight distribution according to season')
plt.show() 
```

# Histogram

* A histogram is a representation of the distribution of grouped data using a bar chart.
* It involves transferring data consisting of repeated numbers to a table after the applied operations, and then using the table to create a graph, that is, showing the graph of data groups in the form of rectangular bars.

```python
sns.displot(olympicData, x = 'height')
plt.ylabel('Frequency')
plt.title('Histogram of Height')
plt.show() 
```

# Histogram - Categorical

```python
sns.displot(olympicData, x = 'height', hue = 'gender')
plt.ylabel('Frequency')
plt.title('Histogram of Height')
plt.show() 
```

```python
sns.displot(olympicData, x = 'height', col = 'gender', multiple = 'dodge')

plt.show() 
```

# 2D Histogram

```python
sns.displot(olympicData, x = 'height', y = 'weight', kind = 'kde')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Histogram of Height & Weight')
plt.show() 
```

```python
sns.displot(olympicData, x = 'height', y = 'weight', hue = 'gender', kind = 'kde')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Histogram of Height and weight distribution according to genders')
plt.show() 
```

```python
sns.kdeplot(data = olympicData, x = 'height', y = 'weight', hue = 'gender', fill = True)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Histogram of Height and weight distribution according to genders')
plt.show() 
```

# Bar Plot

* A bar chart or bar graph is a graph that presents categorical data with rectangular bars with heights or lengths proportional to the values they represent.

```python
sns.barplot(x = 'medal', y = 'age', data = olympicData)
plt.title('Medal distribution of Height')
plt.show()
```

```python
sns.barplot(x = 'medal', y = 'age', data = olympicData, hue = 'gender')
plt.title('Medal distribution of Height')
plt.show()
```

```python
sns.catplot(x = 'medal', y = 'age', hue = 'gender', col = 'season', kind = 'bar',
            height = 5, aspect = 0.7, data = olympicData)
plt.show()
```

# Box Plot

* In statistics, a box plot is a descriptive statistical and graphical tool developed as an explanatory data analysis tool to visually summarize numerical data in a box-and-whisker plot format. (Information about the box plot is provided in the data science section.)

```python
sns.boxplot(x = 'season', y = 'height', data = olympicData)
plt.show()
```

```python
sns.boxplot(x = 'season', y = 'height', hue = 'gender', palette = 'rocket', data = olympicData)
plt.show()
```

```python
oly = olympicData.loc[:, ['age','weight', 'height']]
```

```python
sns.boxplot(orient = 'h', palette = 'Set1', data = oly)
plt.show()
```

```python
sns.catplot(x = 'season', y = 'height', hue = 'gender', col = 'medal', kind = 'box',
            height = 5, aspect = 0.7, data = olympicData)
```

# Heat Map

* A heat map is a technique used to visually represent the relationship between numerical features/columns.

```python
oly = olympicData.loc[:, ['age','weight', 'height', 'year']]
```

```python
sns.heatmap(oly.corr(), annot = False, linewidths = 0.5, fmt = '.1f')
plt.show()
```

* We used a correlation matrix to examine the relationship between the numerical features of height, weight, age, and years.
* As we already know, there is a positive relationship of 0.8 between height and weight.

# Violin Graph

* A violin plot is a method for plotting numerical data.
* It is interpreted similarly to a box plot.

```python
sns.violinplot(x = 'season', y = 'height', data = olympicData)
plt.show()
```

```python
sns.violinplot(x = 'season', y = 'height', hue = 'gender', data = olympicData)
plt.show()
```

```python
sns.violinplot(x = 'season', y = 'height', hue = 'gender', data = olympicData, split = True)
plt.show()
```

```python
sns.catplot(x = 'season', y = 'height', hue = 'gender', col = 'medal', kind = 'violin',
            height = 5, aspect = 0.7, data = olympicData, split = True)
```

# Joint Plot
* It is a plot used to examine the relationship between different features using various visualization methods.

```python
sns.jointplot(data= olympicData, x = 'weight', y = 'height', hue = 'season', kind = 'kde')
plt.show()
```

```python
g = sns.JointGrid(data= olympicData, x = 'weight', y = 'height')
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

# Pair Plot
* It is a plot used to examine the relationship between numerical features within a dataset using various visualization methods.

```python
sns.pairplot(olympicData)
plt.show()
```

```python
g = sns.PairGrid(olympicData)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill = True)
g.map_diag(sns.histplot, kde = True)
plt.show()
```

# Count Plot
* A count plot can be thought of as a categorical histogram instead of a numerical variable.

```python
sns.countplot(x = 'city', data = olympicData)
plt.xticks(rotation = 90)
plt.show()
```

* The Olympics have been held in different cities.
* We can see how many Olympics were held in each city.
